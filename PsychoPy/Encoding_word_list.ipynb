```python

# Function to generate word pairs with different first letters
import random

#TODO: Change algorithm so it Draws the words from one df using conditions for emo / ima high not to draw from separate lists 

# Function to generate word pairs with different first letters from two different DataFrames
def generate_word_pairs(df1, column_name1, df2, column_name2, num_pairs, seed=42):
    # Set seed for reproducibility
    random.seed(seed)
    
    word_pairs = []
    words1 = list(zip(df1[column_name1], [df1.name] * len(df1)))  # Add DataFrame name to each word
    words2 = list(zip(df2[column_name2], [df2.name] * len(df2)))  # Add DataFrame name to each word
    
    #error handling when requested words are to few in the input list
    if len(words1) < num_pairs:
        raise ValueError(f"Number of words in pairs exceeds number of words in list 1")
    if len(words2) < num_pairs:
        raise ValueError(f"Number of words in pairs exceeds number of words in list 2")
    
    #initalize sets
    used_words1 = set()
    used_words2 = set()
    
    while len(word_pairs) < num_pairs:
        
        word1, df1_name = random.choice(words1)  # Draw randomly from df1
        word2, df2_name = random.choice(words2)  # Draw randomly from df2
        
        # Check if both words are not already used in any word pair
        if (word1, df1_name) not in used_words1 and (word2, df2_name) not in used_words2:
            word_pairs.append(((word1, df1_name), (word2, df2_name)))
            used_words1.add((word1, df1_name))
            used_words2.add((word2, df2_name))
            
    # Extracting individual elements from inner tuples
    words1, dfs1 = zip(*[pair[0] for pair in word_pairs])
    words2, dfs2 = zip(*[pair[1] for pair in word_pairs])

    # Creating a DataFrame
    df = pd.DataFrame({
        'Word1': words1,
        'List1': dfs1,
        'Word2': words2,
        'List2': dfs2
    })
    
   #shuffel rows randomly
    #shuffled_df = df.sample(frac=1, random_state=seed).reset_index(drop=True)
    
    #return shuffled_df
    return df

generate_word_pairs(emo_high,"WORD", ima_high, "WORD", 12)
#print(generate_word_pairs(emo_high,"WORD", ima_high, "WORD", 10))
```

```python

# Function to generate word pairs with different first letters
import random

# Function to generate word pairs with different first letters from two different DataFrames
def generate_word_pairs(df1, column_name1, df2, column_name2, num_pairs, seed=42):
    # Set seed for reproducibility
    random.seed(seed)
    
    word_pairs = []
    words1 = list(zip(df1[column_name1], [df1.name] * len(df1)))  # Add DataFrame name to each word
    words2 = list(zip(df2[column_name2], [df2.name] * len(df2)))  # Add DataFrame name to each word
    
    #error handling when requested words are to few in the input list
    if len(words1) < num_pairs:
        raise ValueError(f"Number of words in pairs exceeds number of words in list 1")
    if len(words2) < num_pairs:
        raise ValueError(f"Number of words in pairs exceeds number of words in list 2")
    
    #set counters
    counter_df1 = 0
    counter_df2 = 0
    
    #initalize sets
    used_words1 = set()
    used_words2 = set()
    
    while len(word_pairs) < num_pairs:
        # Determine from which data frame to draw
        if counter_df1 <= num_pairs // 2:
            word1, df1_name = random.choice(words1)  # Draw randomly from df1
            word2, df2_name = random.choice(words2)  # Draw randomly from df2
            counter_df1 += 1
        else:
            word1, df1_name = random.choice(words2)  # Draw randomly from df2
            word2, df2_name = random.choice(words1)  # Draw randomly from df1
            counter_df2 += 1
        
        # Check if both words are not already used in any word pair
        if (word1, df1_name) not in used_words1 and (word2, df2_name) not in used_words2:
            word_pairs.append(((word1, df1_name), (word2, df2_name)))
            used_words1.add((word1, df1_name))
            used_words2.add((word2, df2_name))
            
    # Extracting individual elements from inner tuples
    words1, dfs1 = zip(*[pair[0] for pair in word_pairs])
    words2, dfs2 = zip(*[pair[1] for pair in word_pairs])

    # Creating a DataFrame
    df = pd.DataFrame({
        'Word1': words1,
        'List1': dfs1,
        'Word2': words2,
        'List2': dfs2
    })
    
   #shuffel rows randomly
    #shuffled_df = df.sample(frac=1, random_state=seed).reset_index(drop=True)
    
    #return shuffled_df
    return df

generate_word_pairs(emo_high,"WORD", ima_high, "WORD", 12)
#print(generate_word_pairs(emo_high,"WORD", ima_high, "WORD", 10))
```
