```python
#define function
def shuffle(word_list):
    #extract lines 3 to 37 
    rows_for_shuffeling = nature_words[3:39]
    primacy_words = nature_words[:3]
    recency_words = nature_words[39:]

    #shuffel 
    seed = 42
    shuffled_rows = rows_for_shuffeling.sample(frac=1, random_state=seed).reset_index(drop=True)
    shuffled_df = pd.concat([primacy_words,shuffled_rows, recency_words], ignore_index=True)
    
    return shuffled_df
```
