# Experiment 2 - Rated Word Pair list

-->  Results of Experiment 1 (generation of semantic pairs) and Experiment 2 (rating of strength)

Columns exp-ratings:
- with the target relation, 
- the target word, 
- the generated response, 
- the number of ratings for a strength of 0/1/2/3/4/5, and 
- the number of ratings indicating the number of "I don't know" cases.

```python

# File paths
path1 = "/Users/Richard/PycharmProjects/NeuroLab_3.10/WordList/sem-rel-database/exp-rating/N_ratings_asANT.txt"
path2 = "/Users/Richard/PycharmProjects/NeuroLab_3.10/WordList/sem-rel-database/exp-rating/N_ratings_asHYP.txt"
path3 = "/Users/Richard/PycharmProjects/NeuroLab_3.10/WordList/sem-rel-database/exp-rating/N_ratings_asSYN.txt"

# Read CSV files
ratings1 = pd.read_csv(path1, sep="\t", header=None)
ratings2 = pd.read_csv(path2, sep="\t", header=None)
ratings3 = pd.read_csv(path3, sep="\t", header=None)

# Define column names
columns = ["relation", "target", "response", "N_Strength_0", "N_Strength_1", "N_Strength_2", "N_Strength_3", "N_Strength_4", "N_Strength_5", "N_Idk_cases"]

# Concatenate DataFrames with column names assigned
all_ratings = pd.concat([ratings1, ratings2, ratings3], ignore_index=True)
all_ratings.columns = columns

# Display the first few rows
#all_ratings[all_ratings["relation"] != "ANT"].head(50)
```

## Merge two lists SemRel and EmoIma 

--> 

```python
#Which item in semrel is also in noun
common_target = all_ratings[all_ratings["target"].isin(Emo["WORD"])]
common_response = noun[noun["response"].isin(Emo["WORD"])]
common_pairs = all_ratings[all_ratings["target"].isin(Emo["WORD"]) & all_ratings["response"].isin(Emo["WORD"])]

# delete duplicate target words (first occurence of the word is kept)
unique_common_pairs = common_pairs.drop_duplicates(subset=['target'])

#unique_common_pairs.head(30)
unique_common_pairs.head(50)
```

```python
common_pairs["relation"].value_counts()
unique_common_pairs["relation"].value_counts()
```

```python
print(f'Number of targets in EmoIma: {len(common_target)}')
print(f'Number of responses in EmoIma: {len(common_response)}')
print(f'Number of common Noun-Noun pairs: {len(common_pairs)}')
print(f'Number of common unique Noun-Noun pairs: {len(unique_common_pairs)}')
```
